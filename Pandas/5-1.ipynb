## 分析哪一個職業的男性比率最高
* 將性別轉為數字
* 函式Apply( )
* 新增DataFrame 欄位
* 計算出每一個職業中男生所佔的比例，並且降冪排序(公式：每一個職業男性數量的總和 ÷ 每一個職業的總數)

### 建立一個函式，將性別轉為數字

```python
def gender_to_numeric(x):
    if x== 'M':
        return 1
    if x == 'F':
        return 0
gender_to_numeric('M')
```

### 函式 Apply
將特定函式應用在DataFrame中，不需要用傳統的for迴圈方式，處理資料更為方便<br>
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.apply.html

```python
users['gender'].apply(gender_to_numeric)
```

### 新增DataFrame 欄位

```python
users['gender_n'] = users['gender'].apply(gender_to_numeric)
users
```

### 計算出每一個職業中男生所佔的比例，並且降冪排序
公式：每一個職業男性數量的總和 ÷ 每一個職業的總數

```python
result = users.groupby('occupation').gender_n.sum() / users.occupation.value_counts() * 100
result.sort_values(ascending=False)
```

# 計算男性人數

```python
condition=users["gender"]=="M"
men_pop=users[condition]
men_pop
```

```python
men_occ_count=men_pop.groupby("occupation")["gender"].value_counts()
men_occ_count=men_occ_count.unstack()
print(men_occ_count)
print(men_occ_count.columns)
print(men_occ_count.index)
```

```python
men_occ_count.index.name=None
men_occ_count.columns.name=None
men_occ_count
```

```python
men_occ_count=men_pop.groupby("occupation")["gender"].value_counts()
men_occ_count=men_occ_count.groupby("occupation").sum()
men_occ_count.sort_index(ascending=True,inplace=True)
men_occ_count
```

# 計算各職業總人數

```python
total_count=users["occupation"].value_counts()
total_count.sort_index(ascending=True,inplace=True)
total_count
```

# 完成圖表

```python
result=(men_occ_count/total_count)*100
result.sort_values(ascending=False,inplace=True)
result
```
