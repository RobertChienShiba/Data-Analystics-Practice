# 練習
### 餐廳來客數資料分析
資料連結：https://raw.githubusercontent.com/Code-Gym/python-dataset/master/RestaurantVisitors.csv

```python
# import pandas as pd

res = pd.read_csv('https://raw.githubusercontent.com/Code-Gym/python-dataset/master/RestaurantVisitors.csv',sep=',',index_col="date")
res.head()
```

## 將date欄位轉為DatetimeIndex

```python
res.index = pd.to_datetime(res.index)
res
```

# 練習-將date欄位轉為DatetimeIndex

```python
res.index=pd.to_datetime(res.index)
res.head()
```

# 第一間餐廳rest1，每週來客數總和多少？

```python
res["week"]=res.index.isocalendar().week
#res.resample("w")["rest1"].sum()
res.groupby("week")["rest1"].sum()
```

## 假日四間餐廳平均來客數
提示：
* 使用欄位Holiday找出假日資料
* 使用loc過濾四間餐廳的來客數
* 最後算出平均值

```python
condition=(res["holiday"]==1)|(res["weekday"].str.startswith("S"))
res_holiday=res[condition]
res_holiday
```

```python
res_holiday["rest1"].mean()
```

```python
res_holiday.agg({"rest1":"mean","rest2":"mean","rest3":"mean","rest4":"mean"})
```

## 非假日四間餐廳平均來客數

```python
res_workday=res[~condition]
res_workday
```

```python
res_workday.agg({"rest1":"mean","rest2":"mean","rest3":"mean","rest4":"mean"})
```

# 周六和周日的資料

```python
condition_sat=res_holiday["weekday"].str.contains("Saturday")
res_sat=res_holiday[condition_sat]
res_sat
```

```python
res_sat.agg({"rest1":"mean","rest2":"mean","rest3":"mean","rest4":"mean"})
```
